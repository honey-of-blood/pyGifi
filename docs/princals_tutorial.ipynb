# Ordinal and Mixed PCA with Princals

Python equivalent of R Gifi vignette: *Ordinal and Mixed PCA*

**Princals** (Principal Components Analysis with Optimal Scaling) extends classical PCA
to ordinal, nominal and mixed variable types by **optimally scaling** each variable
before decomposition.

---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pygifi import Princals, get_dataset
from pygifi.visualization.plot import plot
```

## 1 · Roskam Personality Scales (Ordinal)

The Roskam dataset (`n=91`) contains personality scale data measured on Likert-type ordinal scales.

```python
roskam = get_dataset('roskam')
print(roskam.shape)
print(roskam.head())
print('\nValue ranges:')
print(roskam.describe().loc[['min', 'max']])
```

### 1.1 · Fit Princals (ordinal)

```python
p_roskam = Princals(ndim=2, ordinal=True, eps=1e-8, itmax=1000).fit(roskam)
print(p_roskam)
```

### 1.2 · Loadings Plot

```python
fig, ax = plt.subplots(figsize=(7, 6))
plot(p_roskam, plot_type='loadplot', ax=ax, title='Roskam — Component Loadings')
plt.tight_layout()
plt.show()
```

### 1.3 · Transformation Plot (transplot)

For each variable, plots the **original ordinal rank** vs. the **optimally scaled value** in dimension 1.

```python
fig, ax = plt.subplots(figsize=(9, 6))
plot(p_roskam, plot_type='transplot', ax=ax, title='Roskam — Optimal Transformations')
plt.tight_layout()
plt.show()
```

### 1.4 · Biplot

Combines object scores and loading vectors in a single plot. Loading vectors are scaled by $\sqrt{n}$.

```python
fig, ax = plt.subplots(figsize=(8, 7))
plot(p_roskam, plot_type='biplot', ax=ax, title='Roskam — Biplot')
plt.tight_layout()
plt.show()
```

### 1.5 · Summary Statistics

```python
p_roskam.summary()
```

## 2 · Neumann Dataset (Mixed: Metric + Ordinal)

The Neumann dataset has numeric and ordinal predictors — a typical **mixed measurement** scenario.

```python
neumann = get_dataset('neumann')
print(neumann.shape)
print(neumann.head())
```

```python
# Treat all as ordinal
p_neumann = Princals(ndim=2, ordinal=True, eps=1e-8, itmax=1000).fit(neumann)
print(p_neumann)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
plot(p_neumann, plot_type='loadplot', ax=axes[0], title='Neumann — Loadings')
plot(p_neumann, plot_type='transplot', ax=axes[1], title='Neumann — Transformations')
plt.tight_layout()
plt.show()
```

## 3 · Copies: Multiple Quantifications per Variable

The `copies` parameter allows a variable to contribute multiple quantifications (multi-dimensional optimal scaling per variable).

```python
from pygifi import get_dataset
galo3 = get_dataset('galo').iloc[:, 0:3]  # 3-column subset

# Three variables; second variable gets 2 copies
p_copies = Princals(ndim=2, copies=[1, 2, 1], eps=1e-8, itmax=1000).fit(galo3)
print(p_copies)

# The transforms for variable 1 have 2 columns (one per copy)
tr1 = p_copies.result_['transform'][1]
print(f'\nTransform shape for variable 1 (copies=2): {tr1.shape}')
```
