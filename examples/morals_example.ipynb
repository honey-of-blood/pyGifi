# PyGifi: Morals Example (Multiple Optimal Regression)
This notebook demonstrates `Morals`. Unlike standard OLS which draws a straight line through metric data, `Morals` attempts to *optimally scale* both the input predictors ($X$) and the target variable ($y$) non-linearly to maximize the total $R^2$ of the regression.

```python
import numpy as np
import pandas as pd
from pygifi import Morals

pd.set_option('display.max_columns', None)
```

### 1. Generate a Mixed Dataset
We will generate a fake dataset with two categorical predictors (`A`, `B`, `C` and `Low`, `Med`, `High`) and a continuous target bounded between 0 and 10.

```python
np.random.seed(42)

# Create categorical predictors
X = pd.DataFrame({
    'Category_1': np.random.choice(['A', 'B', 'C'], 100),
    'Category_2': np.random.choice(['Low', 'Med', 'High'], 100)
})

# Create target variable
y = np.random.rand(100) * 10

print("Predictors (X):")
print(X.head(3))
print("\nTarget (y):")
print(y[:3])
```

### 2. Fit the Morals Model
We instantiate Morals, declaring the scaling level for each variable axis. 
*   `ordinal_y=True` means the continuous target `y` will be warped monotonically to find a better fit.
*   `ordinal_x=[False, True]` means predictor 1 (`Category_1`) is treated as entirely nominal, while predictor 2 (`Category_2`) is ordered (`Low` < `Med` < `High`).

```python
# Instruct the model on scaling levels for the input array
model = Morals(ordinal_x=[False, True], ordinal_y=True)

# Standard Scikit-Learn Signature
model.fit(X, y)
```

### 3. Review Regression Results
Let's observe the generated optimal predictions and final coefficients. The `beta_` array contains the linear regression weights applied to the *optimally scaled* versions of our categorical parameters.

```python
print(f"Optimized R-Squared: {model.r_squared_:.4f}")
print(f"Linear Beta Coefficients: {model.beta_}")
```

### 4. Visualization
A key component of understanding a Gifi regression is seeing exactly *how* the variables were transformed to achieve that $R^2$. Pygifi provides `.plot_transplot_y()` specifically to show how the target variable was warped.

```python
import matplotlib.pyplot as plt

# Plot Target optimal scaling mapping
model.plot_transplot_y()

# Plot transformations applied to Predictor 2 (The ordinal variable)
model.plot_transplot(variable='Category_2')
```
