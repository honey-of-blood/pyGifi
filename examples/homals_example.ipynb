# PyGifi: Homals Example (Multiple Correspondence Analysis)
This notebook demonstrates how to load a dataset and fit a Homogeneity Analysis by Alternating Least Squares (`Homals`) model to it. Homals operates exclusively on **nominal** categorical data.

```python
import pandas as pd
from pygifi import Homals
from pygifi.data import datasets

# Set display options for better visibility
pd.set_option('display.max_columns', None)
```

### 1. Load the Data
We will use the built-in `hartigan` dataset, which describes the characteristics of several hardware components using categorical names.

```python
df = datasets.get_dataset('hartigan')
df.head()
```

### 2. Fit the Homals Model
We instantiate Homals to capture 2 dimensions. Homals will assign optimal coordinates to each category to maximize the explained variance.

```python
model = Homals(ndim=2, eps=1e-6)
model.fit(df)
```

### 3. Review the Results
Let's look at the captured variance (eigenvalues) and the final stress value.

```python
print(f"Final Stress: {model.stress_:.4f}")
print(f"Eigenvalues (Variance Captured per Dim): {model.eigenvalues_}")
```

We can also look at the exact numeric quantifications assigned to a specific categorical variable. Notice how PyGifi automatically handles string categories and assigns arrays for both dimensions.

```python
print("Quantifications for 'Thread':")
print(model.category_quantifications_['Thread'])
```

### 4. Visualization
Finally, we can visualize the projected object scores (the rows) and category quantifications in the new 2D geometric space. Pygifi has built in matplotlib wrappers for this.

```python
import matplotlib.pyplot as plt

# 1. Plot the row scores
model.plot_objectscores()

# 2. Plot the category mappings
model.plot_quantifications()
```
