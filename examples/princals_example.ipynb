# PyGifi: Princals Example (Non-linear PCA)
This notebook demonstrates Principal Components Analysis by Alternating Least Squares (`Princals`). Unlike Homals, Princals allows constraints on the transformations—for example, enforcing *ordinal* rules using Monotone/Isotonic Regression so that ordered categories maintain their relative ranks.

```python
import pandas as pd
from pygifi import Princals
from pygifi.data import datasets

pd.set_option('display.max_columns', None)
```

### 1. Load the Data
We will use the built-in `neumann` dataset. This involves weather and physical characteristics measured over multiple days, presenting a mix of data that is ideally treated ordinally.

```python
df = datasets.get_dataset('neumann')
df.head()
```

### 2. Fit the Ordinal Princals Model
Here we instruct the model to assume *all* columns are ordinal (`ordinal=True`). Ties are strictly maintained (`ties='s'`), and we will ask the library to use multiple imputation via ALS for any missing cells (`missing='m'`).

```python
model = Princals(ndim=2, ordinal=True, ties='s', missing='m', eps=1e-5)
model.fit(df)
```

### 3. Review the Results
Eigenvalues define the variance successfully extracted by the non-linear PCA.

```python
print(f"Final Stress: {model.stress_:.4f}")
print(f"Eigenvalues: {model.eigenvalues_}")
```

### 4. Visualization: Non-linear Transformation Plots
The unique strength of Princals is mathematically warping standard metrics or ranks into their mathematically optimal shape *without breaking the required ordering*. Let's plot the "Transformation curve" for a column.

```python
import matplotlib.pyplot as plt

# Look at how the original 'Wind' values were mathematically bent (step-function due to ties='s')
model.plot_transplot(variable='Wind')

# Look at how 'Temperature' was ordinally scaled
model.plot_transplot(variable='Temperature')
```

Like Homals, Princals also supports geometric visualization of Object Scores and Loadings (the Principal Components).

```python
# Plot the variable loadings
model.plot_loadings()
```
